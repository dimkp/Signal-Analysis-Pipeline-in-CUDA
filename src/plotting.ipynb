```python
import numpy as np
import matplotlib.pyplot as plt

signal = np.loadtxt("unfiltered_output.txt")
output = np.loadtxt("filtered_output.txt")

plt.plot(signal, label="Original")

plt.legend()
plt.title("Signal Signal")
plt.show()


plt.plot(output, label="Filtered")

plt.legend()
plt.title("Filtered Signal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("frequency_filtered_output.txt")
freq = data[:, 0]
mag = data[:, 1]

plt.plot(freq, mag)
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```
